## Desarrollo del reto
Aplica funciones y an√°lisis seg√∫n el objetivo. Usa `def` para definir tus funciones.

```python
# PASO 2: Crear la funci√≥n resumen_termico()

def resumen_termico(df):
    resumen = pd.DataFrame({
        'temp_promedio': [df['temperatura'].mean()],
        'temp_max': [df['temperatura'].max()],
        'temp_min': [df['temperatura'].min()],
        'hum_promedio': [df['humedad'].mean()],
        'hum_max': [df['humedad'].max()],
        'hum_min': [df['humedad'].min()]
    })
    return resumen

# Ejecutamos la funci√≥n y mostramos el resultado
resumen = resumen_termico(df)
display(resumen)
```

```python
# PASO 3: Crear la funci√≥n clasificar_clima()

def clasificar_clima(temp, hum):
    if temp >= 18 and hum >= 70:
        return 'h√∫medo templado'
    elif temp >= 18 and hum < 70:
        return 'seco templado'
    elif temp < 18 and hum >= 70:
        return 'h√∫medo fr√≠o'
    else:
        return 'seco fr√≠o'

# Probar la funci√≥n con un ejemplo
ejemplo = clasificar_clima(20, 80)
print("Ejemplo: temp=20, hum=80 ‚Üí", ejemplo)
```

```python
# PASO 4: Aplicar la funci√≥n a cada fila del DataFrame

df['clima'] = df.apply(lambda x: clasificar_clima(x['temperatura'], x['humedad']), axis=1)

# Mostrar las primeras filas para verificar que se agreg√≥ correctamente
display(df.head())
```

```python
# PASO 5: Crear la tabla de frecuencias por tipo de clima

tabla_frecuencias = df['clima'].value_counts().reset_index()
tabla_frecuencias.columns = ['Tipo de clima', 'N√∫mero de sitios']

display(tabla_frecuencias)
```

```python
# PASO 6: Guardar los resultados en archivos CSV

# Guardar el DataFrame completo con la nueva columna "clima"
df.to_csv("resultados_clasificados.csv", index=False)

# Guardar la tabla de frecuencias por tipo de clima
tabla_frecuencias.to_csv("resumen_climas.csv", index=False)

print("‚úÖ Archivos guardados correctamente:")
print("- resultados_clasificados.csv")
print("- resumen_climas.csv")
```
