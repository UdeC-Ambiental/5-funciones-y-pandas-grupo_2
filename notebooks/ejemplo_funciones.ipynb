## 1Ô∏è‚É£ ¬øQu√© es una funci√≥n?

Una funci√≥n es un bloque de c√≥digo reutilizable que se ejecuta cuando es llamado. Se define con `def` y puede recibir **par√°metros** y devolver un **resultado** con `return`.

```python
def saludar():
    print('Hola, bienvenido al laboratorio de Python Ambiental!')

saludar()
```

```python
def suma(x, y):
    return x+y

resultado = suma(4,5)
print('La suma de 4 mas 5 es:', resultado)
```

## 2Ô∏è‚É£ Funciones con condicionales

Podemos combinar condicionales dentro de una funci√≥n para tomar decisiones seg√∫n los valores recibidos.

```python
def clasificar_ph(ph):
    if ph < 6.5:
        return '√Åcido'
    elif ph <= 8.5:
        return 'Neutro'
    else:
        return 'B√°sico'

print(clasificar_ph(6.0))
print(clasificar_ph(7.4))
print(clasificar_ph(9.1))
```

## 3Ô∏è‚É£ Funciones con bucles

Podemos usar bucles dentro de funciones para procesar colecciones de datos (listas o rangos num√©ricos).

```python
def promedio(lista):
    suma = 0
    for valor in lista:
        suma += valor
        print(suma)
    return suma / len(lista)

temperaturas = [18.2, 21.5, 19.8, 22.3]
print('Promedio de temperatura:', round(promedio(temperaturas), 2))
```

## 4Ô∏è‚É£ Funciones con estructuras de almacenamiento

Ahora aplicamos funciones para analizar datos ambientales en diferentes estructuras.

```python
def analizar_temperaturas(lista_tmeperatura):
    for t in lista_tmeperatura:
        if t > 25:
            print(t, '‚Üí Alta temperatura')
        else:
            print(t, '‚Üí Temperatura normal')

# Lista
temperaturas = [18.5, 22.4, 27.1]
analizar_temperaturas(temperaturas)

# Tupla
mediciones = (19.0, 25.3, 29.5)
analizar_temperaturas(mediciones)
```
