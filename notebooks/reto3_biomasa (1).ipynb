```python
import pandas as pd 
df = pd.read_csv("resultados_experimentos.csv")
df.head()
```

```python
#mostrar filas y tipos de datos
df.info()
```

```python
#estadisticas descriptivas
df.describe()
```

```python
#verificar si hay valores nulos
df.isnull().sum()
```

```python
#vamos a clasificar la biomasa de los ecosistemas segun su nivel
def clasificar_biomasa(valor):
    if valor < 50:
        return "baja"
    elif valor < 150:
        return "media"
    else:
        return "alta"
      
```

```python
#aplicamos la funcion
df["categoria_biomasa"] = df["biomasa"].apply(clasificar_biomasa)
```

```python
#se cuenta cuantos ecosistemas hay por cada categoria
df["categoria_biomasa"].value_counts()
```

```python
#promedio y analisis de los ecosistemas al ver cuanta biomasa tienen
df.groupby("categoria_biomasa").mean(numeric_only=True)
```
