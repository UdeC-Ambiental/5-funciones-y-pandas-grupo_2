```python
# Paso 2 Desarrollo del reto
# Función que calcule el promedio de biomasa por nivel de luz
def promedio_biomasa_por_luz(df, columna_luz, columna_biomasa):
    return df.groupby(columna_luz)[columna_biomasa].mean()

# Reemplaza los nombres de las columnas según tu archivo:
resultado = promedio_biomasa_por_luz(df, "luz", "biomasa")

# Mostrar el resultado
print(resultado)
```

```python
import matplotlib.pyplot as plt

plt.plot(resultado.index, resultado.values, marker='o')
plt.title("Relación entre Luz y Biomasa")
plt.xlabel("Nivel de Luz")
plt.ylabel("Promedio de Biomasa")
plt.show()
```

```python
# INTERPRETACIÓN DE RESULTADOS
print("Conclusión:")
print("""No parece haber una relación lineal clara entre el nivel de luz y el promedio de biomasa,
ya que los valores de la biomasa fluctúan bastante para diferentes niveles de luz.
En algunos rangos de luz, al aumentar el nivel de luz, aumenta también el promedio de biomasa,
pero en otros rangos desciende o varía abruptamente, lo que sugiere que otros factores pueden estar influyendo en la biomasa,
además de la luz. La dispersión y los cambios bruscos indican alta variabilidad,
lo cual puede deberse a efectos de otros factores ambientales, variabilidad experimental o incluso errores de medición.""")
```
