## Desarrollo del reto
Aplica funciones y análisis según el objetivo. Usa `def` para definir tus funciones.

```python
# Crea aquí tus funciones y análisis
# Función 1 para limpiar los datos
def limpiar_datos(d):
    if d['temperatura'] > 60:
        d['temperatura'] = 60
    if d['humedad'] > 100:
        d['humedad'] = 100
    if d['humedad'] < 0:
        d['humedad'] = 0
    if d['luz'] < 0:
        d['luz'] = 0
    if d['biomasa'] < 0:
        d['biomasa'] = 0
    return d

 # Ejemplo de un sitio con datos extremos
sitio = {'sitio': 'Bosque Alto de Nocaima', 'temperatura': 70, 'humedad': 120, 'luz': 500, 'biomasa': -5}
print('Antes de limpiar:', sitio)

 # Aplicar la función
sitio_limpio = limpiar_datos(sitio)
print('Después de limpiar:', sitio_limpio)



#Función 2 indice de calor
def indice_calor(temp, hum):
    if temp is None or hum is None:
        return None
    resultado = 0.7 * temp + 0.3 * (hum / 5)
    return resultado

 #Ejemplo: calcular el índice de calor para un sitio
temperatura = 28
humedad = 75

 #Aplicar la función
resultado = indice_calor(temperatura, humedad)
print("El índice de calor es:", resultado)

#Función 3 Clasificar ecosistema
def clasificar_ecosistema(temp, hum, luz):
    if temp >= 25 and hum >= 70:
        return "Húmedo-cálido"
    elif hum < 40 and luz > 800:
        return "Seco-soleado"
    else:
        return "Templado"

# Ejemplo para probar
print(clasificar_ecosistema(24, 35, 1200))
```
